```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test(cdata):
    result=adfuller(cdata)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print(" Data has no unit root and is stationary")
    else:
        print(" Data has a unit root, indicating it is non-stationary ")
        
 #If p< 0.05 ; Data is stationary
#if p>0.05; Data is not stationary       

adfuller_test(cdata['No_of_confirmed_cases'])
```

## model

```python
## importing the library
from statsmodels.tsa.arima_model import ARIMA
```

```python
print(cdata2.shape)
```

```python
#Split Your Dataset
train=cdata2.iloc[:220]
test=cdata2.iloc[220:]
print(train.shape,test.shape)
```

```python
train
```

```python
test
```

```python
##Model object creation and fitting the model
model_arima = ARIMA(train, order=(6,1,1))

model_arima_fit=model_arima.fit()

```

```python
## evaluate the model
print(model_arima_fit.aic)
```

## plotting the forecasted values


```python
## plotting the test data
plt.figure(figsize=(15,5))
plt.plot(test)
```

```python
## predicting the out of sample points
predictions = model_arima_fit.forecast(steps=22)[0]
predictions
```

```python
plt.figure(figsize=(15,5))
plt.plot(predictions,color='red')
```

```python
start=len(train)
end=len(train)+len(test)-1
predictions2=model_arima_fit.predict(start=start,end=end,typ='levels')
predictions2.plot(label='test data prediction',legend=True)
test.plot(legend=True)
```
